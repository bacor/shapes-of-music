# Dist-dip

```python
from scipy.spatial import distance_matrix
```

```python
data = np.concatenate([
    np.random.randn(200), 
    np.random.randn(200) + 3,
    np.random.randn(200) + 10
])
dm = distance_matrix(np.array([data]).T, np.array([data]).T)
```

```python
plt.figure(figsize=(10, 3))
plt.subplot(121)
sns.histplot(data, kde=True)
plt.subplot(122)
sns.histplot(dm.ravel(), kde=True)
plt.tight_layout()
```

```python
diptest(np.msort(data), is_hist=False, numt=1000)
```

### 2D example 

```python
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance

data = np.concatenate([
        np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 200),
        np.random.multivariate_normal([6, 0], [[1, 0], [0, 1]], 200),
        np.random.multivariate_normal([3, 8], [[1, 0], [0, 1]], 200)
])
x, y = data.T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
dm = distance_matrix(data, data)
dip, pval, (left_idx, right_idx) = diptest(np.msort(dm.ravel()), is_hist=False, numt=100)
pval
```
