# Figure A: Average phrase contours

Here we plot the average contours in the three corpora

```python
# Import 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import sys
sys.path.append('../')
from helpers import cm2inch, title, show_num_contours, load_datasets
matplotlib.style.use('styles.mplstyle')
```

```python
# Load all data
dfs, contours = load_datasets()
```

```python
def plot_average_contour(contours, baseline, f=.25, color='C3'):
    xs = np.linspace(0, 1, contours.shape[1])
    b_mean = baseline.mean(axis=0)
    b_std = (f/2) * baseline.std(axis=0)
    plt.plot(xs, b_mean, 'k', lw=1, zorder=-1, alpha=.5, label='baseline')
    plt.fill_between(xs, b_mean-b_std, b_mean+b_std, alpha=.1, color='k',lw=0)
    
    mean = contours.mean(axis=0)
    std = f/2 * contours.std(axis=0)
    plt.plot(xs, mean, color, lw=1, label=f'phrases')
    plt.fill_between(xs, mean-std, mean+std, alpha=.2, color=color, label=f'${f:.2f}\sigma$', lw=0)
    plt.plot([0, 1], [0, 0], 'k:', lw=.5, alpha=.5)

# plt.figure(figsize=cm2inch(8,4))
# plot_average_contour(contours['erk'], contours['erk-random'])
```

```python
plt.figure(figsize=cm2inch(12,3.5))
ax1 = plt.subplot(131)
plot_average_contour(contours['erk'], contours['erk-random'], color='C0')
title('Western: Erk')
show_num_contours(len(contours['erk']), ax1)
plt.ylabel('normalized pitch (st)')
plt.xlabel('position')
plt.ylim(-2.5, 2)

ax2 = plt.subplot(132, sharey=ax1)
plot_average_contour(contours['liber-antiphons'], contours['liber-antiphons-random'], color='C1')
show_num_contours(len(contours['liber-antiphons']), ax2)
title('Gregorian: antiphons')
plt.xlabel('position')
plt.legend()

ax3 = plt.subplot(133, sharey=ax1)
plot_average_contour(contours['han'], contours['han-random'], color='C2')
show_num_contours(len(contours['han']), ax3)
title(f'Chinese: Han')
plt.xlabel('position')

plt.tight_layout()
plt.savefig('../figures/figA1-average-contours-subset.pdf')
# plt.savefig('../figures/fig01-average-contours.png', dpi=150)
```

## All datasets

```python
title = lambda x: plt.title(x, ha='left', x=0)
def show_num_contours(num_contours, ax, y=1.06):
    plt.text(1, y, f'N={num_contours}', transform=ax.transAxes, 
         va='bottom', ha='right', size=6, fontstyle='italic', color='0.6')

    
plt.figure(figsize=cm2inch(12,9))
ax1 = plt.subplot2grid((3, 3), (0, 0), rowspan=1, colspan=1)
plot_average_contour(contours['erk'], contours['erk-random'], color='C0')
title('Western: Erk')
show_num_contours(len(contours['erk']), ax1)
plt.ylabel('normalized pitch (st)')

ax2 = plt.subplot2grid((3, 3), (0, 1), sharey=ax1)
plot_average_contour(contours['boehme'], contours['boehme-random'], color='C0')
show_num_contours(len(contours['boehme']), ax2)
title(f'Boehme')

ax3 = plt.subplot2grid((3, 3), (0, 2), sharey=ax1)
plot_average_contour(contours['creighton'], contours['creighton-random'], color='C0')
show_num_contours(len(contours['creighton']), ax3)
title('Creighton')

ax4 = plt.subplot2grid((3, 3), (1, 0))
plot_average_contour(contours['han'], contours['han-random'], color='C1')
title('Chinese: Han')
show_num_contours(len(contours['han']), ax4)
plt.ylabel('normalized pitch (st)')


ax5 = plt.subplot2grid((3, 3), (1, 1), sharey=ax4)
plot_average_contour(contours['shanxi'], contours['shanxi-random'], color='C1')
show_num_contours(len(contours['shanxi']), ax5)
title('Shanxi')

ax6 = plt.subplot2grid((3, 3), (1, 2), sharey=ax4)
plot_average_contour(contours['natmin'], contours['natmin-random'], color='C1')
show_num_contours(len(contours['natmin']), ax6)
title('Natmin')

ax7 = plt.subplot2grid((3, 3), (2, 0))
plot_average_contour(contours['liber-antiphons'], contours['liber-antiphons-random'], color='C2')
show_num_contours(len(contours['liber-antiphons']), ax7)
title('Gregorian: Antiphons')
plt.xlabel('position')
plt.ylabel('normalized pitch (st)')

ax8 = plt.subplot2grid((3, 3), (2, 1), sharey=ax7)
plot_average_contour(contours['liber-alleluias'], contours['liber-alleluias-random'], color='C2')
show_num_contours(len(contours['liber-alleluias']), ax8)
title('Alleluias')
plt.xlabel('position')

ax9 = plt.subplot2grid((3, 3), (2, 2), sharey=ax7)
plot_average_contour(contours['liber-responsories'], contours['liber-responsories-random'], color='C2')
show_num_contours(len(contours['liber-responsories']), ax9)
title('Responsories')
plt.xlabel('position')
plt.legend()

plt.tight_layout()
plt.savefig('../figures/figA2-average-contours')
```
