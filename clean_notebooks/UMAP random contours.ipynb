```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import umap
import umap.plot

import sys
sys.path.append('../')
from helpers import title

from src.clusterability.dataset import Dataset
from src.clusterability.visualization import show_contour_grid
```

```python
def show_embeddings(mapper):
    fig, ax = plt.subplots(1, figsize=(10,10))
    points = mapper.embedding_[:, :]
    x, y = points[:2000, :].T
    plt.scatter(x, y, s=5, alpha=.4)
    show_contour_grid(mapper.inverse_transform, points, num_samples=20, eps=.5)
```

### Markov

```python
dataset = Dataset('markov')
contours = dataset.representation('pitch_centered', limit=5000)
mapper = umap.UMAP().fit(contours)
```

```python
show_embeddings(mapper)
```

### Binom

```python
dataset = Dataset('binom')
contours = dataset.representation('pitch_centered', limit=5000)
mapper_binom = umap.UMAP().fit(contours)
```

```python
show_embeddings(mapper_binom)
```
