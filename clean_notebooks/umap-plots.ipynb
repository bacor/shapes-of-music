## Side plot

### Illustrate grid construction

```python
from scipy.spatial import ConvexHull
from matplotlib.path import Path
from itertools import product
from matplotlib.gridspec import GridSpec
```

```python
mapper = markov_mapper
umap.plot.points(mapper)

# parameters
margin = 0.05
subdivisions=15

# Compute complex hull
points = mapper.embedding_
hull = ConvexHull(points)
hull_path = Path(points[hull.vertices])

# Visualize complex hull
ax = plt.gca()
for simplex in hull.simplices:
    ax.plot(points[simplex, 0], points[simplex, 1], '.--C1')

# Grid in box around dataset
xmin, ymin = hull.min_bound
xmax, ymax = hull.max_bound
width = xmax - xmin
height = ymax - ymin
xmargin = margin * height
ymargin = margin * width
xgrid = np.linspace(xmin - xmargin, xmax + xmargin, subdivisions)
ygrid = np.linspace(ymin - ymargin, ymax + ymargin, subdivisions)
gridpoints = np.array(list(product(xgrid, ygrid)))

# Determine gridpoints inside/outside the convex hull
radius = margin * min(height, width)
inside = hull_path.contains_points(gridpoints, radius=radius)
outside = inside == False

x, y = gridpoints.T
plt.scatter(x[inside], y[inside], marker='x', c='k', s=20)
plt.scatter(x[outside], y[outside], marker='x', c='0.8', s=20)
```

```python
# Now using grid_around_points function
umap.plot.points(mapper)
points = mapper.embedding_
gridpoints, inside = grid_around_points(points, subdiv=20)
x, y = gridpoints.T
plt.scatter(x[inside], y[inside], marker='x', c='k', s=20, lw=1)
```

## Real dataset: combined

```python
combined = Dataset('combined-phrase')
```

### Intervals

```python
interval_contours = combined.representation('interval', limit=2500)
interval_mapper = umap.UMAP(random_state=0).fit(interval_contours)

gridpoints, inside = grid_around_points(interval_mapper.embedding_, subdiv=10, margin=0.1)
mean_inv_contours = average_inv_contours(interval_mapper.inverse_transform, gridpoints, num_samples=2)
mean_inv_contours = mean_inv_contours.cumsum(axis=1)
```

```python
fig = plt.figure(figsize=(12, 6))
show_side_plot(interval_mapper, gridpoints, inside, mean_inv_contours,
    umap_plot_kws=dict(labels=combined.subset_column('dataset', limit=2500))
)
```

### Pitch representation

The pitch representation primarily sorts contours based on their initial pitch, as expected

```python
pitch_contours = combined.representation('pitch', limit=2000)
pitch_mapper = umap.UMAP(random_state=0).fit(pitch_contours)

gridpoints, inside = grid_around_points(pitch_mapper.embedding_, subdiv=10, margin=0.1)
mean_inv_contours = average_inv_contours(pitch_mapper.inverse_transform, gridpoints, num_samples=2)
```

```python

fig = plt.figure(figsize=(12, 6))
show_side_plot(pitch_mapper, gridpoints, inside, mean_inv_contours,
    umap_plot_kws=dict(
        labels=combined.subset_column('0', limit=2000), 
        cmap='magma',
        show_legend=False
    )
)
```
