```python
import music21
```

```python
s = music21.converter.parse('../datasets/boehme/deut2272.krn')
```

```python
def analyze_tonic(stream):
    aarden = stream.analyze('Aarden')
    krumhansl = stream.analyze('Krumhansl')
    final = stream.flat.notes[-1]
    analysis = {
        'tonic_aarden': aarden.tonic.ps,
        'tonic_krumhansl': krumhansl.tonic.ps,
        'tonic_final': final.pitch.ps,
    }
    analysis['tonic_agreement'] = aarden.tonic.name == krumhansl.tonic.name == final.name
    return analysis

analyze_tonic(s)
```
